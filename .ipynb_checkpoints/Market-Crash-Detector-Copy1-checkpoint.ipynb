```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.metrics import confusion_matrix, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import statsmodels.api as sm
```

```python
data = pd.read_csv("MarketData/TWII_Daily.csv") 
data = data.dropna()
```

```python
plt.plot(data.Close)
```

```python
data['Return'] = data.Close.pct_change()
data.head()
```

```python
max_slope = 1 + 60
max_vol = 1 + 60
for i in range(1, max_slope):
    data['r'+str(i)] = data.Close.pct_change(i)
for i in range(5, max_vol):
    data['vol'+str(i)] = data.Close.rolling(i).std()
data
```

```python
data = data.dropna().iloc[100:,:].reset_index().iloc[:,1:]
data
```

```python
from scipy.signal import argrelextrema
```

```python
n= 20 # number of points to be checked before and after 
# Find local peaks
reg_target = pd.DataFrame()
reg_target['Date'] = data.Date
reg_target['Close'] = data.Close
reg_target['min'] = data.iloc[argrelextrema(data.Close.values, np.less_equal, order=n)[0]]['Close']
reg_target['max'] = data.iloc[argrelextrema(data.Close.values, np.greater_equal, order=n)[0]]['Close']
```

```python
reg_target['is_min'] = (reg_target['min'] > 0).astype(int) * -1 
reg_target['is_max'] = (reg_target['max'] > 0).astype(int) * 1 
reg_target['target'] = reg_target.is_min + reg_target.is_max
```

```python
reg_target = reg_target.dropna(axis=1)
reg_target
```

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

regr2 = RandomForestRegressor(n_estimators=100, max_depth=30)
```

```python
train_end = 4000
test_point = 1000
X_train = data.iloc[:train_end, 8:]
y_train = reg_target['target'].iloc[:train_end]
X_test = data.iloc[train_end:train_end+test_point, 8:]
y_test = reg_target['target'].iloc[train_end:train_end+test_point]

price_train = data.Close.iloc[:train_end].values
price_test = data.Close.iloc[train_end:train_end+test_point].values
```

```python
regr2.fit(X_train, y_train)
```

```python
regr2.score(X_train, y_train)
```

```python
regr2.score(X_test, y_test)
```

```python
def scale_output(x, top=None, bot=None):
    if top == None :
        top = abs(max(x))
    if bot == None : 
        bot = abs(min(x))
    for i in range(len(x)):
        if x[i] > 0 :
            x[i] = x[i]/top
        elif x[i] < 0 :
            x[i] = x[i]/bot
    return x, top, bot 
```

```python
pred_train = regr2.predict(X_train)
pred_test = regr2.predict(X_test)

pred_test, top, bot= scale_output(pred_test)
# pred_test = scale_output(pred_test, top, bot)[0]

# pred_train = np.reshape(pred_train, (-1,1))
# pred_test = np.reshape(pred_test, (-1,1))
# sc = MinMaxScaler(feature_range=(-1,1))
# sc = StandardScaler( with_mean=False, with_std=True)
# sc.fit(pred_train)
# pred_train = sc.transform(pred_train)
# pred_test = sc.transform(pred_test)
# pred_train = np.reshape(pred_train, pred_train.shape[0])
# pred_test = np.reshape(pred_test, pred_test.shape[0])
pred_test[:20]
```
