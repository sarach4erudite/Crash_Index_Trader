```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_finance import candlestick_ohlc
import ta 

try : 
    from keras.models import Sequential, model_from_json
    from keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense, Dropout, LSTM
    from keras.callbacks import TensorBoard, EarlyStopping
    from keras.metrics import categorical_accuracy, mean_absolute_percentage_error
except : 
    from tensorflow.keras.models import Sequential, model_from_json
    from tensorflow.keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense, Dropout, LSTM
    from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
    from tensorflow.keras.metrics import categorical_accuracy, mean_absolute_percentage_error
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('Data/EURUSD_4H_NEW.csv')
csv_result = "Result/MyResult_LSTM_EURUSD.csv"

predict_skip = 5 
rolling_untrend = 6*5

data['Date'] = pd.to_datetime(data.Date)
data['Year'] = data.Date.dt.year
shift_return = 5 
data['Return'] = data.Close.pct_change(shift_return).shift(-shift_return)

indicator_length = 6*5
data['RSI_Short'] = ta.rsi(data.Close, n=indicator_length)/100
data['RSI_Long'] = ta.rsi(data.Close, n=indicator_length*2)/100
data['Volatility_Short'] = data.Close.rolling(indicator_length).std()
data['Volatility_Long'] = data.Close.rolling(indicator_length*2).std()
data['Close_DPO_Short'] = ta.dpo(data.Close, n=indicator_length)
data['Close_DPO_Long'] = ta.dpo(data.Close, n=indicator_length*2)
data['WillR_Short'] = ta.wr(data.High, data.Low, data.Close, indicator_length)/100+1 
data['WillR_Long'] = ta.wr(data.High, data.Low, data.Close, indicator_length*2)/100+1


data = data.dropna()

data = data[(data.Year >= 2011) & (data.Year <= 2018)]

#data = data.iloc[-100:]

input_data = pd.DataFrame() 
input_data['Date'] = data.Date
input_data['Time'] = data.Time
input_data['Close_DPO_Short']= data['Close_DPO_Short']
input_data['Close_DPO_Long']= data['Close_DPO_Long']
input_data['RSI_Short'] = data['RSI_Short']
input_data['RSI_Long'] = data['RSI_Long']
input_data['Volatility_Short'] = data['Volatility_Short']
input_data['Volatility_Long'] = data['Volatility_Long']
input_data['WillR_Short'] = data['WillR_Short']
input_data['WillR_Long'] = data['WillR_Long']

class_bound = {} 
class_num = 5
class_range = 1/(class_num)

print("Labeling Data ...")
for i in range(1,class_num+1):
    if i != class_num : 
        class_bound[i] = data.Return.quantile(class_range*i)
    
#plt.hist(data.Return, bins='auto',color='blue')
#for i in range(1, class_num):    
#    plt.axvline(class_bound[i], color='red')
target = data[['Date']]
for i in range(class_num):  
    if i == 0 :
        data['Class_'+str(i)] = (data.Return <= class_bound[i+1]).astype(int)
        target['Class_'+str(i)] = data['Class_'+str(i)]
    elif i == class_num-1 : 
        data['Class_'+str(i)] = (data.Return > class_bound[i]).astype(int) 
        target['Class_'+str(i)] = data['Class_'+str(i)]
    else : 
        data['Class_'+str(i)] = ((data.Return > class_bound[i]) & (data.Return <= class_bound[i+1])).astype(int)
        target['Class_'+str(i)] = data['Class_'+str(i)]

short_candles_window = 6*5
long_candle_window= 6*10

print("Divide Train/Test Set ... ")
train_length = int(len(input_data)*0.9)
print("Train Start : ", input_data.Date.iloc[long_candle_window])
print("Train End : ", input_data.Date.iloc[train_length-1])
print("Test Start :", input_data.Date.iloc[train_length])
print("Test End :", input_data.Date.iloc[-1])

sc = MinMaxScaler()
sc.fit(data[['Close_DPO_Short']].iloc[:train_length])
input_data['Close_DPO_Short']= sc.transform(input_data[['Close_DPO_Short']])

sc.fit(data[['Close_DPO_Short']].iloc[:train_length])
input_data['Close_DPO_Long']= sc.transform(input_data[['Close_DPO_Long']])

sc.fit(data[['Volatility_Short']].iloc[:train_length])
input_data['Volatility_Short'] = sc.transform(input_data[['Volatility_Short']])

sc.fit(data[['Volatility_Long']].iloc[:train_length])
input_data['Volatility_Long'] = sc.transform(input_data[['Volatility_Long']])
```

```python
x_train = []
y_train = [] 
x_test = [] 
y_test = []

start_index = long_candle_window
for i in range(start_index, len(input_data)):
#     print("Create Image at Date : ", input_data.Date.iloc[i], "Time :", input_data.Time.iloc[i])
    x_input = input_data.iloc[i-long_candle_window:i, 2:].values
    y_target = target.iloc[i,1:].values
    if i < train_length :
        x_train.append(x_input)
        y_train.append(y_target)
    else : 
        x_test.append(x_input)
        y_test.append(y_target)

x_train = np.array(x_train) 
x_test = np.array(x_test) 
y_train = np.array(y_train) 
y_test = np.array(y_test) 

print('Create Data Done ...')
# model_input_shape = (x_train.shape[1], x_train.shape[2], x_train.shape[3])
```

```python
# x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1], 1))
# x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1], 1))
print('input shape : ', x_train.shape)
model_input_shape = (x_train.shape[1], x_train.shape[2])
```

```python
def test_model(model, x_test, y_test, date=[], prices=[], init_profit=1):
    p = model.predict(x_test)

    true_class = [] 
    for i in range(len(y_test)):
        true_class.append(np.argmax(y_test[i]))
    pred_class = []     
    for i in range(len(p)):
        pred_class.append(np.argmax(p[i]))
        
    # true_class = pd.Series(true_class)
    # pred_class = pd.Series(pred_class)
    return pred_class, true_class
```

```python
def train_model(model, x_train, y_train, epochs=130):
    tbCallBack = TensorBoard(log_dir='./Graph')
    es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=100)
    model.fit(x_train, y_train, batch_size=256, epochs=epochs, validation_split=0.05, shuffle=False, callbacks=[tbCallBack, es])
    # model.fit(x_train, y_train, batch_size=32, epochs=epochs, shuffle=True)
    return model 
```

```python
def init_model():
    model = Sequential()

    model.add(LSTM(128, input_shape=model_input_shape,return_sequences=True, activation='relu'))
    model.add(Dropout(0.5))
    model.add(LSTM(32, activation='relu'))
    model.add(Dropout(0.5))
#     model.add(LSTM(16, activation='relu'))
#     model.add(Dropout(0.5))
#     model.add(Dense(128, activation='relu'))
#     model.add(Dropout(0.5))

    model.add(Dense(class_num, activation='softmax'))

#     model.compile(optimizer='adam', loss='mse', metrics=[categorical_accuracy])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=[categorical_accuracy])
    return model 
```

```python
def save_model(model, name=''):
    model_json = model.to_json()
    with open("Weight/"+name+".json", "w") as json_file:
        json_file.write(model_json)
    # serialize weights to HDF5
    model.save_weights("Weight/"+name+".h5")
        # serialize model to YAML
    model_yaml = model.to_yaml()
    with open("Weight/"+name+".yaml", "w") as yaml_file:
        yaml_file.write(model_yaml)
    # serialize weights to HDF5
    print("Saved model to disk")
```

```python
print("Input shape : ", model_input_shape)
epoch = 600

print("Training model ...")
model_name = "train_lstm_model"
model = init_model() 
model.summary()
model = train_model(model, x_train, y_train, epochs=epoch)
save_model(model, model_name)
result = pd.DataFrame()

pred_class, true_class = test_model(model, x_test, y_test)

predicted_class = pred_class
real_class = true_class
test_date = data.Date.iloc[train_length:]
test_time = data.Time.iloc[train_length:]
test_price = data.Close.iloc[train_length:]

result['Date'] = test_date
result['Time'] = test_time
result['Close'] = test_price 
result['y_pred'] = predicted_class
result['y_true'] = real_class

try : 
    result.to_csv(csv_result, index=False)
    print("Saved result to csv")
except : 
    pass
```

```python
cm = confusion_matrix(result.y_pred, result.y_true)
plt.figure(figsize=(300,300))
plt.imshow(cm, cmap=plt.cm.Blues)
plt.show()
```
